```python
# Calculate percentage of missing values per column
missing_percentage = (regional_data.isnull().sum() / len(regional_data)) * 100

# Filter columns with missing values
missing_columns = missing_percentage[missing_percentage > 0].sort_values(ascending=False)

# Display missing values summary
print("Columns with Missing Values (Percentage):")
print(missing_columns)

# Visualize missing values
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
missing_columns.plot(kind='bar', color='orange', alpha=0.7)
plt.title('Percentage of Missing Values per Column', fontsize=16)
plt.xlabel('Columns', fontsize=12)
plt.ylabel('Percentage (%)', fontsize=12)
plt.xticks(rotation=90, fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Key Columns with High Missing Values:

```python
# Compare columns with potential redundancy
redundancy_check = regional_data[['fr_orig', 'dms_orig', 'fr_dest', 'dms_dest', 'fr_inmode', 'fr_outmode', 'dms_mode']]

# Show unique value counts for each column
for col in redundancy_check.columns:
    print(f"Unique values in {col}: {regional_data[col].nunique()}")

# Check correlation between potentially redundant columns
correlation = redundancy_check.corr()
print("\nCorrelation Matrix:\n", correlation)
```

Analysis of Redundancy Check Results
Unique Value Counts:

fr_orig and fr_dest: Only 8 unique values each, compared to 132 unique values in dms_orig and dms_dest.
Likely, fr_orig and fr_dest represent broader region categories, while dms_orig and dms_dest provide detailed region codes.
fr_inmode and fr_outmode: Only 7 unique values, while dms_mode has 8 unique values.
The higher granularity in dms_mode suggests that fr_inmode and fr_outmode may be redundant.
Correlation Matrix:

fr_orig vs. dms_orig: Weak negative correlation (-0.125), indicating these columns are not directly aligned and may represent different levels of regional aggregation.
fr_dest vs. dms_dest: Weak negative correlation (-0.137), with similar implications as above.
fr_inmode and fr_outmode vs. dms_mode:
Moderate to strong positive correlation (0.59 and 0.62, respectively), suggesting that dms_mode captures much of the information in fr_inmode and fr_outmode.
Insights:
fr_orig and fr_dest:

These columns are likely aggregated region categories and might be useful for summary-level insights (e.g., trade flows between broader regions).
Since dms_orig and dms_dest are more granular, we may not need both.
fr_inmode and fr_outmode:

These columns are likely redundant since dms_mode provides a more comprehensive representation of the mode of transportation.
Unless additional context is provided, these columns can likely be dropped.


```python
regional_data_cleaned = regional_data.drop(columns=['fr_inmode', 'fr_outmode'])
```

```python
# Drop redundant columns
regional_data_cleaned = regional_data.drop(columns=['fr_inmode', 'fr_outmode'])

# Verify the cleaned dataset
print("Remaining Columns After Cleanup:")
print(regional_data_cleaned.columns)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Countplot for modes of transport
plt.figure(figsize=(10, 6))
sns.countplot(data=regional_data_cleaned, x='dms_mode', palette='Set2')
plt.title('Distribution of Modes of Transport', fontsize=16)
plt.xlabel('Mode of Transport', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Purpose of the Bar Chart:

This chart shows the distribution of modes of transport used across all regions.
Each bar represents the count of entries for a particular mode of transport.
Actionable Insights:

Dominant Transport Modes: Identify which modes (e.g., rail, truck, air) dominate the freight movement.
Balance of Transport Modes: See whether the dataset is heavily skewed towards one or two modes, or if there's a fairly even distribution of transport types.


```python
# Countplot for trade type (domestic vs international)
plt.figure(figsize=(8, 6))
sns.countplot(data=regional_data_cleaned, x='trade_type', palette='Set1')
plt.title('Distribution of Trade Types (Domestic vs International)', fontsize=16)
plt.xlabel('Trade Type', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(ticks=[0, 1], labels=['Domestic', 'International'], fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
