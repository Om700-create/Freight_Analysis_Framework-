```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
years = [2012, 2013, 2014, 2015, 2020, 2025, 2030, 2035, 2040, 2045]
freight_volume = [1.7e7, 1.75e7, 1.8e7, 1.82e7, 2.05e7, 2.3e7, 2.4e7, 2.5e7, 2.6e7, 2.7e7]
freight_value = [1.75e7, 1.78e7, 1.83e7, 1.86e7, 2.2e7, 2.5e7, 2.75e7, 3.05e7, 3.5e7, 3.75e7]

# Trendlines (using polynomial fitting)
volume_trend = np.poly1d(np.polyfit(years, freight_volume, deg=2))
value_trend = np.poly1d(np.polyfit(years, freight_value, deg=2))

# Plotting
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot Freight Volume
ax1.plot(years, freight_volume, marker='o', linestyle='--', color='teal', label='Freight Volume (tons)')
ax1.plot(years, volume_trend(years), color='teal', linestyle='-', alpha=0.5, label='Volume Trendline')
ax1.set_xlabel('Year')
ax1.set_ylabel('Total Volume (in tons)', color='teal')
ax1.tick_params(axis='y', labelcolor='teal')

# Second Y-axis for Freight Value
ax2 = ax1.twinx()
ax2.plot(years, freight_value, marker='^', linestyle='--', color='orange', label='Freight Value (USD)')
ax2.plot(years, value_trend(years), color='orange', linestyle='-', alpha=0.5, label='Value Trendline')
ax2.set_ylabel('Total Value (in USD)', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Grid and Titles
ax1.grid(visible=True, linestyle='--', alpha=0.5)
plt.title('Freight Growth Trends: Volume vs Value (2012–2045)')

# Legends
fig.legend(loc="upper left", bbox_to_anchor=(0.1, 0.9), fontsize=9)

# Show Plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
years = [2012, 2013, 2014, 2015, 2020, 2025, 2030, 2035, 2040, 2045]
freight_volume = [1.7e7, 1.75e7, 1.8e7, 1.82e7, 2.05e7, 2.3e7, 2.4e7, 2.5e7, 2.6e7, 2.7e7]
freight_value = [1.75e7, 1.78e7, 1.83e7, 1.86e7, 2.2e7, 2.5e7, 2.75e7, 3.05e7, 3.5e7, 3.75e7]

# Trendlines
volume_trend = np.poly1d(np.polyfit(years, freight_volume, 2))
value_trend = np.poly1d(np.polyfit(years, freight_value, 2))

# Creating subplots
fig, axes = plt.subplots(2, 1, figsize=(10, 10), sharex=True)

# Plot 1: Freight Volume
axes[0].plot(years, freight_volume, 'o--', color='teal', label='Freight Volume')
axes[0].plot(years, volume_trend(years), '-', color='teal', alpha=0.7, label='Volume Trendline')
axes[0].annotate(f'{freight_volume[-1]/1e7:.2f}M tons', xy=(2045, freight_volume[-1]), 
                 xytext=(2035, freight_volume[-1]*0.95), arrowprops=dict(arrowstyle='->', color='teal'))
axes[0].set_ylabel('Total Volume (in tons)', color='teal')
axes[0].grid(visible=True, linestyle='--', alpha=0.5)
axes[0].legend()
axes[0].set_title('Freight Volume Over the Years')

# Plot 2: Freight Value
axes[1].plot(years, freight_value, '^--', color='orange', label='Freight Value')
axes[1].plot(years, value_trend(years), '-', color='orange', alpha=0.7, label='Value Trendline')
axes[1].annotate(f'${freight_value[-1]/1e7:.2f}M', xy=(2045, freight_value[-1]), 
                 xytext=(2035, freight_value[-1]*0.95), arrowprops=dict(arrowstyle='->', color='orange'))
axes[1].set_ylabel('Total Value (in USD)', color='orange')
axes[1].grid(visible=True, linestyle='--', alpha=0.5)
axes[1].legend()
axes[1].set_title('Freight Value Over the Years')

# Final Adjustments
plt.xlabel('Year')
plt.tight_layout()
plt.show()
```

This visualization separates freight volume and freight value into two subplots for improved clarity.

Top Subplot: Freight Volume

Freight Volume is plotted in teal using a dashed line with markers.
A trendline (solid line) smoothly approximates the growth of freight volume over the years.
A highlighted annotation at 2045 (2.70M tons) draws attention to the final projected value.
Bottom Subplot: Freight Value

Freight Value is plotted in orange with triangular markers and dashed lines.
A trendline (solid line) represents the growth trajectory of freight value.
An annotation highlights the final value of $3.75M USD in 2045.
Layout:

Subplots are vertically aligned with shared x-axes for easy comparison across years.
Titles, gridlines, and color-coded legends ensure clarity.


```python
import matplotlib.pyplot as plt
import numpy as np

# Data
years = [2012, 2015, 2020, 2025, 2030, 2035, 2040, 2045]
freight_volume = [1.7e7, 1.82e7, 2.05e7, 2.3e7, 2.4e7, 2.5e7, 2.6e7, 2.7e7]
freight_value = [1.75e7, 1.86e7, 2.2e7, 2.5e7, 2.75e7, 3.05e7, 3.5e7, 3.75e7]

# Trendlines
volume_trend = np.poly1d(np.polyfit(years, freight_volume, 2))
value_trend = np.poly1d(np.polyfit(years, freight_value, 2))

# Create the plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot Freight Volume (Primary Y-axis)
ax1.plot(years, freight_volume, 'o--', color='teal', label='Freight Volume')
ax1.plot(years, volume_trend(years), '-', color='teal', alpha=0.7, label='Volume Trendline')
ax1.set_ylabel('Total Volume (in tons)', color='teal')
ax1.tick_params(axis='y', labelcolor='teal')
ax1.grid(visible=True, linestyle='--', alpha=0.5)

# Annotate Key Volume Years
for i, year in enumerate([2020, 2030, 2045]):
    ax1.annotate(f'{freight_volume[i+2]/1e7:.2f}M tons', xy=(year, freight_volume[i+2]),
                 xytext=(year-5, freight_volume[i+2]*0.95),
                 arrowprops=dict(arrowstyle='->', color='teal'))

# Add a Secondary Y-axis for Freight Value
ax2 = ax1.twinx()
ax2.plot(years, freight_value, '^--', color='orange', label='Freight Value')
ax2.plot(years, value_trend(years), '-', color='orange', alpha=0.7, label='Value Trendline')
ax2.set_ylabel('Total Value (in USD)', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Annotate Key Value Years
for i, year in enumerate([2020, 2030, 2045]):
    ax2.annotate(f'${freight_value[i+2]/1e7:.2f}M', xy=(year, freight_value[i+2]),
                 xytext=(year+2, freight_value[i+2]*1.05),
                 arrowprops=dict(arrowstyle='->', color='orange'))

# Add Vertical Lines for Decade Markers
for year in [2020, 2030, 2045]:
    plt.axvline(x=year, color='gray', linestyle='--', alpha=0.5)

# Title and Legend
fig.suptitle('Freight Growth Trends: Volume vs Value with Decade Projections', fontsize=14)
fig.tight_layout()
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))
plt.show()
```
