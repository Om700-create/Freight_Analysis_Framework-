```python
# Calculate percentage of missing values per column
missing_percentage = (regional_data.isnull().sum() / len(regional_data)) * 100

# Filter columns with missing values
missing_columns = missing_percentage[missing_percentage > 0].sort_values(ascending=False)

# Display missing values summary
print("Columns with Missing Values (Percentage):")
print(missing_columns)

# Visualize missing values
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
missing_columns.plot(kind='bar', color='orange', alpha=0.7)
plt.title('Percentage of Missing Values per Column', fontsize=16)
plt.xlabel('Columns', fontsize=12)
plt.ylabel('Percentage (%)', fontsize=12)
plt.xticks(rotation=90, fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Summing up total freight volume for each year
freight_volume = {
    '2012': regional_data_cleaned['tons_2012'].sum(),
    '2013': regional_data_cleaned['tons_2013'].sum(),
    '2014': regional_data_cleaned['tons_2014'].sum(),
    '2015': regional_data_cleaned['tons_2015'].sum(),
    '2020': regional_data_cleaned['tons_2020'].sum(),
    '2025': regional_data_cleaned['tons_2025'].sum(),
    '2030': regional_data_cleaned['tons_2030'].sum(),
    '2035': regional_data_cleaned['tons_2035'].sum(),
    '2040': regional_data_cleaned['tons_2040'].sum(),
    '2045': regional_data_cleaned['tons_2045'].sum(),
}

# Convert to DataFrame for easy plotting
freight_volume_df = pd.DataFrame(list(freight_volume.items()), columns=['Year', 'Total Volume'])

# Plotting the trend
plt.figure(figsize=(10, 6))
plt.plot(freight_volume_df['Year'], freight_volume_df['Total Volume'], marker='o', color='teal', linestyle='-')
plt.title('Total Freight Volume Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Volume (in tons)')
plt.grid(True)
plt.show()
```

```python
# Summing up total freight value for each year
freight_value = {
    '2012': regional_data_cleaned['value_2012'].sum(),
    '2013': regional_data_cleaned['value_2013'].sum(),
    '2014': regional_data_cleaned['value_2014'].sum(),
    '2015': regional_data_cleaned['value_2015'].sum(),
    '2020': regional_data_cleaned['value_2020'].sum(),
    '2025': regional_data_cleaned['value_2025'].sum(),
    '2030': regional_data_cleaned['value_2030'].sum(),
    '2035': regional_data_cleaned['value_2035'].sum(),
    '2040': regional_data_cleaned['value_2040'].sum(),
    '2045': regional_data_cleaned['value_2045'].sum(),
}

# Convert to DataFrame for easy plotting
freight_value_df = pd.DataFrame(list(freight_value.items()), columns=['Year', 'Total Value'])

# Plotting the trend
plt.figure(figsize=(10, 6))
plt.plot(freight_value_df['Year'], freight_value_df['Total Value'], marker='o', color='darkorange', linestyle='-')
plt.title('Total Freight Value Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Value (in USD)')
plt.grid(True)
plt.show()
```

Explanation of the Output 📊
Chart Title: "Total Freight Value Over the Years"

This chart displays how the total freight value (in USD) has changed across various years (from 2012 to 2045).
X-axis (Year):

Represents the years from 2012 to 2045. The years included are:
2012, 2013, 2014, 2015, 2020, 2025, 2030, 2035, 2040, 2045.
Y-axis (Total Value in USD):

The total freight value is represented in USD with a scale of 10^7 (10 million).
For example:
~1.75e7 = $17.5 million (in 2012)
~3.75e7 = $37.5 million (in 2045).
Trend:

The chart shows a steady increase in freight value over time.
Between 2012 and 2015, the growth is slower (~1.75e7 to ~1.9e7).
From 2020 onwards, the growth becomes more significant and continues to accelerate until 2045.
Insights:

The sharp rise in value after 2020 suggests increased economic activity, trade, or infrastructure improvements.
By 2045, the freight value has more than doubled compared to 2012.


```python
# Plotting both Freight Volume and Value on the same chart
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plotting Total Volume
ax1.plot(freight_volume_df['Year'], freight_volume_df['Total Volume'], marker='o', color='teal', linestyle='-', label='Total Volume (tons)')
ax1.set_xlabel('Year')
ax1.set_ylabel('Total Volume (in tons)', color='teal')
ax1.tick_params(axis='y', labelcolor='teal')

# Creating a second Y-axis for Total Value
ax2 = ax1.twinx()
ax2.plot(freight_value_df['Year'], freight_value_df['Total Value'], marker='o', color='darkorange', linestyle='-', label='Total Value (USD)')
ax2.set_ylabel('Total Value (in USD)', color='darkorange')
ax2.tick_params(axis='y', labelcolor='darkorange')

# Title and Grid
plt.title('Comparison of Total Freight Volume and Value Over the Years')
fig.tight_layout()
plt.grid(True)
plt.show()
```
